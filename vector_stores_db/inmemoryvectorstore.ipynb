```python
embeddings_model = OpenAIEmbeddings()
```

```python
embeddings_model.embed_query("The king is the husband of the queen who is his wife").__len__()
```

```python
documents = [globals()[f"document_{i}"] for i in range(1, 11)]
```

```python
vectorstore = InMemoryVectorStore(embedding=embeddings_model)
```

```python
vectorstore.add_documents(documents)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={
    "k": 4
})
```
